```python
import pandas as pd
```

```python
df_nf = pd.read_csv("data/nifty50.csv")
df_bnf = pd.read_csv("data/banknifty.csv")
df_vix = pd.read_csv("data/indiavix.csv")
df_fts = pd.read_csv("data/starter_features.csv")
```

```python
display(df_nf.head())
display(df_bnf.head())
display(df_vix.head())
display(df_fts.head())
```

```python
display(df_nf.info())
display(df_bnf.info())
display(df_vix.info())
display(df_fts.info())
```

```python
# Conversion to pandas dataframe for uniformity

df_nf["date"] = pd.to_datetime(df_nf["date"])
df_bnf["date"] = pd.to_datetime(df_bnf["date"])
df_vix["date"] = pd.to_datetime(df_vix["date"])
df_fts["date"] = pd.to_datetime(df_fts["date"])
```

```python
# renaming duplicate columns so that they dont coincide/conflict after merging

df_nf = df_nf.add_prefix("nf_")
df_bnf = df_bnf.add_prefix("bnf_")
df_vix = df_vix.add_prefix("vix_")


# keeping date same for all dataframes

df_nf.rename(columns={"nf_date": "date"}, inplace=True)
df_bnf.rename(columns={"bnf_date": "date"}, inplace=True)
df_vix.rename(columns={"vix_date": "date"}, inplace=True)

# all columns are now uniquely named
```

```python
# merging the dataset
##### merged on dates; how='inner' => intersection of dates, only common dates among the dataframes survive

df_master = (
    df_nf
    .merge(df_bnf, on="date", how="inner")
    .merge(df_vix, on="date", how="inner")
    .merge(df_fts, on="date", how="inner")
)
```

```python
display(df_master.shape)
display(df_master.head())
display(df_master.info())
```

```python
df_master.to_csv("master_dataframe.csv", index=False)
```
